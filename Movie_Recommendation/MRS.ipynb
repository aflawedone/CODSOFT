```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('imdb_top_1000.csv')
```

```python
movie
```

```python
movie = movie[['Series_Title','Genre','Released_Year','Cast','Director','Overview']]
```

```python
movie
```

```python
movie['Genre'] = movie['Genre'].apply(lambda x:x.split(','))
```

```python
movie
```

```python
movie['Cast'] = movie['Cast'].apply(lambda x:x.split(','))
```

```python
movie
```

```python
movie['Overview'] = movie['Overview'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie['Released_Year'] = movie['Released_Year'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie['Director'] = movie['Director'].apply(lambda x:x.split(" "))
```

```python
movie
```

```python
movie['Tags'] = movie['Genre']+movie['Cast']+movie['Released_Year']+movie['Overview']+movie['Director']
```

```python
movie
```

```python
new_df = movie[['Series_Title','Tags']]
```

```python
new_df
```

```python
new_df['Tags']=new_df['Tags'].apply(lambda x: " ".join(x))
```

```python
new_df['Tags'][0]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
new_df['Tags'][1]
```

```python
new_df['Tags'][4]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
ps.stem('loved')
```

```python
ps.stem('dangerous')
```

```python
stem("crime  drama marlon brando al pacino james caan 1972 an organized crime dynasty's aging patriarch transfers control of his clandestine empire to his reluctant son. francis ford coppola")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['Series_Title']== movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].Series_Title)
```

```python
recommend('The Shawshank Redemption')
```
